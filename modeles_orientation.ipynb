```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import NearestNeighbors
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score
from sklearn.impute import SimpleImputer



etudiants = pd.read_csv("TB_ETUDIANT.csv", encoding='latin-1', delimiter=";")

#creation de datafram
df = pd.DataFrame(etudiants)

df
```

```python
# Conversion des variables catégorielles en variables numériques
le_interets = LabelEncoder()
le_comp = LabelEncoder()
df['Interets'] = le_interets.fit_transform(df['Interets '])
df['Compétences'] = le_comp.fit_transform(df['Compétences'])

# Séparation des variables indépendantes (X) et de la variable cible (y)
X = df[['NotesMaths', 'NotesFrançais', 'NotesPhysiques', 'NotesChimie', 'NotesAnglais', 'Interets_1', 'Interets_2']]
y = df['ChoixOrientation']  # Changez cela selon votre cas d'utilisation, par exemple 'ChoixInformatique'


# Division des données en ensembles d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)


# Initialisation du modèle k-NN
knn = KNeighborsClassifier(n_neighbors=3)  # Vous pouvez ajuster le nombre de voisins (k) selon vos besoins

# Entraînement du modèle
knn.fit(X_train, y_train)

# Prédiction sur l'ensemble de test
y_pred = knn.predict(X_test)

# Évaluation du modèle
accuracy = accuracy_score(y_test, y_pred)
print(f"Précision du modèle k-NN : {accuracy}")

```

```python
nouvel_etudiant = {
    'NotesMaths': 20,
    'NotesFrançais': 12,
    'NotesPhysiques': 0,
    'NotesChimie': 0,
    'NotesAnglais': 0,
    'Interets': 'Design',  
    'Compétences': 'Design'  
}

nouvel_etudiant_df = pd.DataFrame([nouvel_etudiant])

nouvel_etudiant_df

# Conversion des variables catégorielles en variables numériques en utilisant les mêmes encodeurs que lors de l'entraînement
nouvel_etudiant_df['Interets'] = le_interets.fit_transform(nouvel_etudiant_df['Interets'])
nouvel_etudiant_df['Compétences'] = le_comp.fit_transform(nouvel_etudiant_df['Compétences'])


# Prédiction avec le modèle k-NN
prediction = knn.predict(nouvel_etudiant_df)

# Affichage de la recommandation
print(f"Recommandation d'orientation pour le nouvel étudiant : {prediction}")

```
