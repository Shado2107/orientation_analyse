```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score



etudiants = pd.read_csv("TB_ETUDIANT.csv", encoding='latin-1', delimiter=";")

#creation de dataframe
df = pd.DataFrame(etudiants)

df
```

```python
df.info()

df.describe().round(2)
```

```python
le_interets = LabelEncoder()
#le_interets2 = LabelEncoder()
le_interets.fit(df['Interets2'])
le_interets.fit(df['Interets1'])


df['Interets2'] = le_interets.fit_transform(df['Interets2'])
df['Interets1'] = le_interets.fit_transform(df['Interets1'])

# Séparation des variables indépendantes (X) et de la variable cible (y)
X = df[['MoyenneBac', 'NotesMaths', 'NotesFrançais', 'NotesPhysChimie', 'NotesSVT', 'NotesAnglais',  'Interets1', 'Interets2']]
y = df['choixOrientation']

# Division des données en ensembles d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)

# Initialisation du modèle k-NN
knn = KNeighborsClassifier(n_neighbors=3)

# Entraînement du modèle
knn.fit(X_train, y_train)

# Prédiction sur l'ensemble de test
y_pred = knn.predict(X_test)

# Évaluation du modèle
accuracy = accuracy_score(y_test, y_pred)
print(f"Précision du modèle k-NN : {accuracy}")
```

```python
nouvel_etudiant = {
    'MoyenneBac' : 15,
    'NotesMaths': 19,
    'NotesFrançais': 19,
    'NotesPhysChimie': 19,
    'NotesSVT': 19,
    'NotesAnglais': 15,
    'Interets1': 'Informatique',  
    'Interets2': 'Biologie'  
}

nouvel_etudiant_df = pd.DataFrame([nouvel_etudiant])

nouvel_etudiant_df

# Conversion des variables catégorielles en variables numériques en utilisant les mêmes encodeurs que lors de l'entraînement
nouvel_etudiant_df['Interets2'] = le_interets.fit_transform(nouvel_etudiant_df['Interets2'])
nouvel_etudiant_df['Interets1'] = le_interets.fit_transform(nouvel_etudiant_df['Interets1'])


# Prédiction avec le modèle k-NN
prediction = knn.predict(nouvel_etudiant_df)

# Affichage des recommandations
print("Recommandations d'orientation pour le nouvel étudiant :")
for i, prediction in enumerate(prediction):
    print(f"Prédiction {i + 1}: {prediction}")

```
